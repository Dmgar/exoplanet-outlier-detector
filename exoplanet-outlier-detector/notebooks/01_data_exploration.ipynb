# 🪐 01 — Exploración de Datos

En este notebook exploramos curvas de luz de estrellas con exoplanetas confirmados,
como punto de partida para entender cómo luce un tránsito real en los datos.

```python
import sys
sys.path.append('..')  # Para importar el paquete desde notebooks/

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from exoplanets.data.loader import LightCurveLoader
from exoplanets.detection.statistical import OutlierDetector
from exoplanets.visualization.plots import plot_lightcurve, plot_score_distribution
```

## 1. Descargar curva de luz de Kepler-11 (sistema con 6 planetas)

```python
loader = LightCurveLoader(mission='Kepler', cadence='long')
lc = loader.fetch('Kepler-11', quarter=1)
print(f'Puntos de datos: {len(lc)}')
lc.head()
```

## 2. Visualización inicial

```python
plot_lightcurve(lc, title='Kepler-11 — Trimestre 1')
```

## 3. Detección de outliers con MAD

```python
detector = OutlierDetector(method='mad', threshold=4.0, direction='down')
mask = detector.fit_predict(lc)

print(f'Outliers detectados: {mask.sum()} de {len(lc)} puntos ({100*mask.mean():.2f}%)')
```

```python
plot_lightcurve(lc, outlier_mask=mask, title='Kepler-11 — Tránsitos Detectados')
```

```python
plot_score_distribution(detector.scores, threshold=4.0)
```

```python
summary = detector.summary(lc, mask)
print(f'Top 10 dips más profundos:')
summary.head(10)
```
